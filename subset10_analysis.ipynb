```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
path = Path(r"C:\Users\rx571gt-b034t\Desktop\PROJECT\subset_10\subset_10_images.csv")
df = pd.read_csv(path, index_col='image_id')
df
```

```python
# le 10 classi che compongono il dataset sono così rappresentate
imgs_per_class = df.value_counts('product_type')
super_classes = imgs_per_class.index.tolist()
listings = df.value_counts('item_id').tolist()
avg_imgs = np.mean(listings)

print(f'Le classi del dataset sono: {", ".join(super_classes)}')
print('\n', imgs_per_class)
print(f'\nNel dataset ci sono {len(listings)} prodotti mediamente con {"%.2f" % avg_imgs} foto ciascuno.')
print(f'L\'inserzione più rappresentata ha {listings[0]} foto.')
```
